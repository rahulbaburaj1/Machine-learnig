```python
import pandas as pd
import numpy as np
df=pd.read_csv("LoanApprovalPrediction.csv")
df
```

```python
df.info
```

```python
df=df.drop('Loan_ID',axis=1)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
x=df['LoanAmount'].mean()
df['LoanAmount'].fillna(x,inplace=True)
```

```python
y=df['Loan_Amount_Term'].mean()
df["Loan_Amount_Term"].fillna(y,inplace=True)
```

```python
z=df['Credit_History'].mean()
df['Credit_History'].fillna(z,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Dependents']=df['Dependents'].str.replace('+','')
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
col=['Gender','Married','Education','Self_Employed','Property_Area']
for i in col:
  df[i]=le.fit_transform(df[i])
```

```python
p=df['Gender'].mean()
df['Gender'].fillna(p,inplace=True)
```

```python
q=df['Married'].mean()
df['Married'].fillna(q,inplace=True)
```

```python
r=df['Dependents'].mode()[0]
df['Dependents'].fillna(r,inplace=True)
```

```python
s=df['Self_Employed'].mean()
df['Self_Employed'].fillna(s,inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
from sklearn import neighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
k_model=KNeighborsClassifier(n_neighbors=7)
n_model=MultinomialNB()
mod=[k_model,n_model]
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
for i in mod:
  print(i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  y_pred
  print("+"*100)
  print(confusion_matrix(y_test,y_pred))
  print("+"*100)
  print(accuracy_score(y_test,y_pred))
```
